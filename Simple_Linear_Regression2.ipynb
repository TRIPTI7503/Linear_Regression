```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
path ="/content/real_estate_price_size.csv"
data = pd.read_csv(path)
print(data)
```

```python
y = data['price']
x1 = data['size']
```

```python
plt.scatter(y,x1)
plt.xlabel('price')
plt.ylabel('size')
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 223.1787 * x1 + 101900
fig = plt.plot(x1,yhat,lw=2,c='k',label='regression line')
plt.xlabel('price')
plt.ylabel('size')
plt.show()
```
